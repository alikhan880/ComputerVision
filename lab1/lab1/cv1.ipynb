```python
import matplotlib.pyplot as plt
import matplotlib.image as pltimg
im = pltimg.imread('.\\6_26.jpg')
fig = plt.figure(figsize=(15,10))

rgb = im
plt.subplot(231)
plt.imshow(rgb)

swprg = im;
swprg[:, :, 1] = im[:, :, 2]
plt.subplot(232)
plt.imshow(swprg)

swprb = im;
swprb[:, :, 1] = im[:, :, 0]
plt.subplot(233)
plt.imshow(swprb)

swpbg = im;
swpbg[:, :, 0] = im[:, :, 2]
plt.subplot(234)
plt.imshow(swpbg)

plt.show()
```
