```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.misc as img
from scipy import ndimage as nimg
import random



def sp_noise(image,prob):
    output = np.zeros(image.shape,np.uint8)
    thres = 1 - prob 
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            rdn = random.random()
            if rdn < prob:
                output[i][j] = 0
            elif rdn > thres:
                output[i][j] = 255
            else:
                output[i][j] = image[i][j]
    return output


im = img.imread("photo_2017-10-13_16-10-28.jpg", mode = 'F')

fig = plt.figure(figsize = (10, 15))

plt.subplot(231)
plt.imshow(im, cmap = 'gray')

noisy = sp_noise(im, 0.01)

plt.subplot(232)
plt.imshow(noisy, cmap = 'gray')

filtered = nimg.median_filter(noisy, 3)
plt.subplot(233)
plt.imshow(filtered, cmap = 'gray')

raznica = filtered - im
raznica2 = im - noisy

plt.subplot(234)
plt.imshow(raznica, cmap = 'gray')

plt.subplot(235)
plt.imshow(raznica2, cmap = 'gray')
plt.show()
```
