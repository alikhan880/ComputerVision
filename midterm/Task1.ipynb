```python
#A
import numpy as np
import matplotlib.pyplot as plt
import scipy.misc as img
from scipy import ndimage as nimg

res = np.array([[1/16, 2/16, 1/16], [2/16, 4/16, 2/16], [1/16, 2/16, 1/16]])
imlen = img.imread('./lena.jpg', mode = 'F')
fig = plt.figure(figsize = (10, 15))
p1 = nimg.convolve(imlen, res, mode='nearest', cval = 0.0)
plt.subplot(221)
plt.imshow(imlen, cmap = 'gray')
plt.subplot(222)
plt.imshow(p1,  cmap = 'gray')
plt.show()

# poorly removes noise such as "salt and pepper"; 


```

```python
#
import numpy as np
import matplotlib.pyplot as plt
import scipy.misc as img
from scipy import ndimage as nimg

imlen = img.imread('./lena.jpg', mode = 'F')
med = nimg.median_filter(imlen, 3)
fig = plt.figure(figsize = (10, 15))
plt.subplot(221)
plt.imshow(imlen, cmap = 'gray')
plt.subplot(222)
plt.imshow(med, cmap = 'gray')
plt.show()


# works slower than the other specified noise-canceling filters;
#increases the sharpness of the image 
```

```python
#
import numpy as np
import matplotlib.pyplot as plt
import scipy.misc as img
from scipy import ndimage as nimg

imlen = img.imread('./lena.jpg', mode = 'F')
arr = np.array([[-1, -2, -1], [-2, 12, -2], [-1, -2, -1]])
res = nimg.convolve(imlen, arr, mode = 'nearest', cval = 0.0)
fig = plt.figure(figsize = (10, 15))

plt.subplot(221)
plt.imshow(imlen, cmap = 'gray')
plt.subplot(222)
plt.imshow(res, cmap = 'gray')
plt.show()

#not a noise-canceling filter
```

```python
#
import numpy as np
import matplotlib.pyplot as plt
import scipy.misc as img
from scipy import ndimage as nimg

imlen = img.imread('./lena.jpg', mode = 'F')
arr = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])
res = nimg.convolve(imlen, arr, mode = 'nearest', cval = 0.0)
fig = plt.figure(figsize = (10, 15))

plt.subplot(221)
plt.imshow(imlen, cmap = 'gray')
plt.subplot(222)
plt.imshow(res, cmap = 'gray')
plt.show()

#not a noise-canceling filter
#will destroy all lines of one pixel thickness in the image
```
