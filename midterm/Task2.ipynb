```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.misc as img
from scipy import ndimage as nimg


im = img.imread("mayk.jpg")

im2 = img.imread("unnamed.png",mode='F')
startIm = im
fig = plt.figure(figsize = (20, 15))
plt.subplot(331)
plt.imshow(startIm)
plt.subplot(332)

plt.hist(im.ravel(), 256, [0, 256])

sumR = 0
sumG = 0
sumB = 0
for i in range (im.shape[0]):
    for j in range(im.shape[1]):
        sumR += im[i, j, 0]
        sumG += im[i, j, 1]
        sumB += im[i, j, 2]
        
meanR = sumR / im.size
meanG = sumG / im.size
meanB = sumB / im.size

avg = (meanR + meanG + meanB) / 3

for i in range(im.shape[0]):
    for j in range(im.shape[1]):
        im[i, j, 0] *= avg / meanR
        im[i, j, 1] *= avg / meanG
        im[i, j, 2] *= avg / meanB
        
        


        
plt.subplot(333)
plt.imshow(im)
plt.show()
```
